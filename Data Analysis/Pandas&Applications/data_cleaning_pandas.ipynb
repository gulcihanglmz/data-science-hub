```python
import pandas as pd
```

```python
df = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Customer Call List.csv")
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.drop("Not_Useful_Column",axis=1,inplace=True)
df
#df.drop(columns = "Not_Useful_Column")
```

```python
#df["Last_Name"]= df["Last_Name"].str.lstrip("...")
#df["Last_Name"]= df["Last_Name"].str.lstrip("/")
#df["Last_Name"]= df["Last_Name"].str.rstrip("_")
df["Last_Name"]= df["Last_Name"].str.strip("123._/")
df
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
df
#  [^a-zA-Z0-9] : harfler ve rakamlar dndaki tm karakterler yerine "" ekle
```

```python
# convert Nan elements to string
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : str(x))
df
```

```python
# #df contains Na&NaN so we must first convert these elements to string
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : x[0:3]+"-" + x[3:6]+"-" + x[6:10])
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace("nan--","")
df["Phone_Number"]
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace("Na--","")
df["Phone_Number"]
```

```python
df
```

```python
df[["Street" , "State" , "Zip_code"]] = df["Address"].str.split(",",expand=True)
df
```

```python
df.head(5)
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"]
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace("No","N")
df["Paying Customer"]
```

```python
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("Yes","Y")
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("No","N")
```

```python
df
```

```python
df = df.replace("N/a","")
```

```python
#df = df.replace("NaN","")
# np.nan veya pd.NA  deerleri replace petodu ile deimez
#NaN" string olarak tanmlanm bir deerdir ve pandas'taki gerek NaN deerleriyle ayn deildir.
#Gerek NaN deerlerini doldurmak veya deitirmek iin fillna() kullanlr
```

```python
df = df.fillna("")
df
```

```python
for x in df.index:
  if df.loc[x,"Do_Not_Contact"] == "Y":    # x:row , Do_Not_Contact : column
    df.drop(x , inplace = True)

df
```

```python
for x in df.index:
  if df.loc[x,"Phone_Number"] == "":
    df.drop(x,inplace=True)

df
```

```python
for x in df.index:
  if df.loc[x,"Do_Not_Contact"] == "":    # x:row , Do_Not_Contact : column
    df.drop(x , inplace = True)

df

```
