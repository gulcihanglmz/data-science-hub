# subset
belli column'lar semek iin "subset" kullanlr

```python
df.dropna(subset=["column1","column2"] , how = "all")
#column1 ve column2'de ikisinde de nan olan deerler silinir
#bir tanesinde nan dieri deilse o satr silinmez
```

```python
df.dropna(subset=["column1","column2"] , how = "any")
#column1 ve column2'de iki satrda birinde bile nan varsa o satr silinir
```

# thresh  = n

```python
df.dropna(thresh = 2)
#en az iki veri varsa o satr silme, o satrda ikiden az deer varsa o satr silinir
```

```python
df.dropna(thresh = 4)  #en az 4 deer varsa o satr silme
```

# Filling the expected values with the average value of the table

```python
print(df.sum())          #her columndaki saylarn toplam
print(df.sum().sum())    #df'deki tm kolonlarn toplam
print(df.size)           #eleman says
print(df.isnull().sum()) #null olan eleman says
print(df.isnull().sum().sum()) #null olan toplam eleman says


def average(df):
  summation = df.sum().sum()
  item_number = (df.size) - (df.isnull().sum().sum())
  avr = summation / item_number
  return avr

result = average(df)
print(f"Average : {result}")

df.fillna(value = average(df))   #nan olan deerleri tablo ortalamasna gre doldurduk
```
