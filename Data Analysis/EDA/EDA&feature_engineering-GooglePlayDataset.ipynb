# Target Encoding (Mean Encoding)

**mean install'a göre genres yerine ort yazmak**


```python
df_clean["Genres"].unique()
```

```python
df_clean.tail()
```

```python
mean_genres_install = df_clean.groupby(["Genres"])["Installs"].mean()/1000000
```

```python
mean_genres_install.reset_index()
```

```python
# df_clean["Genres Encoded"] = mean_genres_install # NaN gelir bu yüzden dict yapıp öyle veriyoruz
```

```python
mean_genres_install = mean_genres_install.to_dict()
```

```python
mean_genres_install  # dict
```

```python
df_clean["Genres Encoded"] = df_clean["Genres"].map(mean_genres_install)
```
